```python
wagon_df['NormalizedWagonType'] = np.where(wagon_df['WAGONTYPE'].str.contains(r'^BCN[0-9A-Z]*$'), 'BCN', wagon_df['WAGONTYPE'].str[:4])
```

```python
wagon_df['NormalizedWagonType'].value_counts()
```

```python
wagon_df['overweight_wagons'] = np.where(wagon_df['ACTUALWT'] > wagon_df['CHRGBLWT']+1, 1,0)
wagon_df['underweight_wagons'] = np.where(wagon_df['ACTUALWT'] < wagon_df['CHRGBLWT'], 1, 0)
wagon_df['optimalweight_wagons']= np.where(wagon_df['ACTUALWT'] == wagon_df['CHRGBLWT'], 1, 0)
```

```python
wagon_df
```

```python
wagon_df['overweight_wagons'].sum(), wagon_df['underweight_wagons'].sum(), wagon_df['optimalweight_wagons'].sum()
```

```python
wagon_df['perc_deadweight'] = wagon_df.apply(lambda row: (row['CHRGBLWT'] - row['ACTUALWT']) / row['CHRGBLWT'] if row['ACTUALWT'] < row['CHRGBLWT'] else 0, axis=1)
```

```python
wagon_df['perc_overweight'] = wagon_df.apply(lambda row: (row['ACTUALWT'] - row['CHRGBLWT']) / row['CHRGBLWT'] if row['ACTUALWT'] > row['CHRGBLWT'] else 0, axis=1)
```

```python
wagon_df
```

```python
wagon_df['capacity_utilisation'] = wagon_df.apply(lambda row: 1 if row['ACTUALWT'] >= row['CHRGBLWT'] else row['ACTUALWT']/row['CHRGBLWT'], axis=1)
```

```python
wagon_df
```

```python
wagon_df.to_excel('wagon_df.xlsx', index= False)
```

```python
grouped_wagon_df = wagon_df.groupby(['INVOICEID']).agg(

    NormalizedWagonType=('NormalizedWagonType', 'first'),

    COMMODITYCODE=('COMMODITYCODE', 'first'),

    WAGONNO=('WAGONNO', 'count'),

    ACTUALWT=('ACTUALWT', 'sum'),

    CHRGBLWT=('CHRGBLWT', 'sum'),

    OVERWTNORM=('OVERWTNORM', 'sum'),

    OVERWTPOL=('OVERWTPOL', 'sum'),

    OVERWTTOTAL=('OVERWTTOTAL', 'sum'),

    overweight_wagons=('overweight_wagons', 'sum'),

    underweight_wagons=('underweight_wagons', 'sum'),

    optimalweight_wagons=('optimalweight_wagons', 'sum'),

    Sum_deadweight=('perc_deadweight', lambda x: (x[x > 0]).sum()),

    Num_deadweight=('perc_deadweight', lambda x: (x[x > 0]).count()),

    Sum_overweight=('perc_overweight', lambda x: (x[x > 0]).sum()),

    Num_overweight=('perc_overweight', lambda x: (x[x > 0]).count()),

    capacity_utilisation=('capacity_utilisation', 'mean')

).reset_index()
```

```python
grouped_wagon_df
```

```python
grouped_wagon_df.to_excel('grouped_wagon_df.xlsx', index= False)
```

```python
# left join of grouped wagon data with the HDR data
hdr_df = pd.read_excel(r"C:\Users\nikhil.upadhyay-v\Desktop\Punitive\RRData\XXHIL_CRIS_RRDATA_HDR_DTL.xlsx")
hdr_df.head()
```

```python
hdr_df.shape
```

```python
#left join with wagon_data
merged_hdr_wagon = pd.merge(hdr_df, grouped_wagon_df, on= 'INVOICEID', how = 'left')
```

```python
merged_hdr_wagon
```

```python
merged_hdr_wagon.rename(columns = {'WAGONNO': 'count_of_wagons', 'overweight_wagons': 'num_of_overweight_wagons',
                                   'underweight_wagons': 'num_of_underweight_wagons', 'optimalweight_wagons': 'num_of_optimalweight_wagons'
                                  }, inplace= True)
```

```python
merged_hdr_wagon['num_of_overweight_rakes'] = np.where(merged_hdr_wagon['ACTLWT'] <= merged_hdr_wagon['CHARGEABLEWEIGHT'], 0, 1)
```

```python
merged_hdr_wagon.head()
```

```python
merged_hdr_wagon['Punitive_weight'] = merged_hdr_wagon['ACTLWT'] - merged_hdr_wagon['CHARGEABLEWEIGHT']
```

```python
merged_hdr_wagon['Punitive_weight_cost'] = merged_hdr_wagon['POLRATE1'] + merged_hdr_wagon['POLRATE2']
```

```python
merged_hdr_wagon.to_csv('merged_hdr_wagon.csv', index= False)
```

Previous Month Calculation

```python
merged_hdr_wagon['INVCDATE'] = pd.to_datetime(merged_hdr_wagon['INVCDATE'], format= '%d-%m-%Y')
```
